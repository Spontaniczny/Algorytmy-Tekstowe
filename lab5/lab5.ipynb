```python
import numpy as np
from PIL import Image
from time import time
```

```python
RED = '\033[91m'
ENDCOLOR = '\033[0m'
```

```python
def count_time(func, *args, **kwargs):
    start_time = time()
    ret = func(*args, **kwargs)
    return time() - start_time, ret
```

# Zad 1

```python
def find_2dtext(text, pattern):
    rows = len(pattern)
    linelength = len(text[0])
    maxlenpattern = 0
    colored_letters = [[False for _ in range(linelength)] for _ in range(len(text))]
    counter = 0
    for p in pattern:
        if len(p) > maxlenpattern:
            maxlenpattern = len(p)
    for i in range(len(text) - rows + 1):
        for j in range(linelength - maxlenpattern + 1):
            x = i
            for p in pattern:
                if p == text[x][j:j+len(p)]:
                    x += 1
                    continue
                break
            else:
                counter += 1
                x = i
                for p in pattern:
                    for xd in range(j, j + len(p)):
                        colored_letters[x][xd] = True
                    x += 1
    for i, line in enumerate(text):
        for j, letter in enumerate(line):
            if colored_letters[i][j]:
                print(f"{RED}{letter}{ENDCOLOR}", end="")
            else:
                print(f"{letter}", end="")
        print()
    print(counter)

def same_letter_next_line(text):
    linelength = len(text[0])
    color_this_line = [False for _ in range(linelength)]
    for i in range(len(text) - 1):
        color_next_line = [False for _ in range(linelength)]
        for j in range(linelength):
            if text[i][j] == text[i + 1][j]:
                color_next_line[j] = True
                print(f"{RED}{text[i][j]}{ENDCOLOR}", end="")
            elif color_this_line[j]:
                print(f"{RED}{text[i][j]}{ENDCOLOR}", end="")
            else:
                print(f"{text[i][j]}", end="")
        print()
        color_this_line = color_next_line
    for j in range(linelength):
        if color_this_line[j]:
            print(f"{RED}{text[-1][j]}{ENDCOLOR}", end="")
        else:
            print(f"{text[-1][j]}", end="")
```

```python
with open("haystack.txt", "r", encoding="UTF-8") as file:
    text = file.readlines()
    maxlen = 0
    for i in range(len(text)):
        text[i] = text[i][:-1]
        if len(text[i]) > maxlen:
            maxlen = len(text[i])
    for i in range(len(text)):
        if len(text[i]) < maxlen:
            x = len(text[i])
            text[i] = text[i] + " " * (maxlen - x)
```

```python
def find_letter_png(image_name, letter):
    tmp = Image.open(image_name).convert("L")
    image = np.array(tmp)
    rows = len(letter)
    imagelength = len(image[0])
    height = len(letter[0])
    colored_letters = [[False for _ in range(imagelength)] for _ in range(len(image))]
    counter = 0
    for i in range(len(image) - rows + 1):
        for j in range(imagelength - height + 1):
            x = i
            for p in letter:
                if np.all(p == image[x][j:j+len(p)]):
                    x += 1
                    continue
                break
            else:
                counter += 1
                x = i
                for p in letter:
                    for xd in range(j, j + len(p)):
                        colored_letters[x][xd] = True
                    x += 1

    tmp = Image.open(image_name)
    image_rgb = np.array(tmp)
    for i in range(len(image_rgb)):
        for j in range(len(image_rgb[0])):
            if colored_letters[i][j] and image[i][j] <= 179:
                image_rgb[i][j] = [255, 0, 0, 255]

    return image_rgb, counter
```

```python
t, ret = count_time(find_letter_png, 'haystack.png', a_array)
image, c = ret
print(f"This took {t} seconds, and {c} letters was found")
display(Image.fromarray(image))
```
