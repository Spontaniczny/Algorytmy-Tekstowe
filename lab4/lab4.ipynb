```python
def distance_levenshtein(a: str, b: str):
    n1, n2 = len(a), len(b)
    tab = [[1 for _ in range(n2+1)] for _ in range(n1+1)]  # tab[a+1][b+1]
    for i in range(n2 + 1):
        tab[0][i] = i
    for i in range(n1 + 1):
        tab[i][0] = i
    for i in range(n1):
        for j in range(n2):
            if a[i] == b[j]:
                tab[i+1][j+1] = tab[i][j]
            else:
                tab[i+1][j+1] = min(tab[i][j + 1], tab[i][j], tab[i + 1][j]) + 1
    # print(tab)

    # for xd in tab:
    #     print(xd)


    # print(tab)

    return tab[-1][-1]


def distance_euclidean(a: str, b: str):
    words_a = re.split(r"[ ,.:;!?'\-()+\"\\<>]", a)
    words_b = re.split(r"[ ,.:;!?'\-()+\"\\<>]", b)
    for i in range(len(words_a)):
        words_a[i] = words_a[i].lower()
    for i in range(len(words_b)):
        words_b[i] = words_b[i].lower()

    all_words = set()
    for word in words_a:
        if word not in all_words:
            all_words.add(word)
    for word in words_b:
        if word not in all_words:
            all_words.add(word)
    dist = 0
    for word in all_words:
        temp = words_a.count(word) - words_b.count(word)
        dist += temp*temp
    # print(dist)
    dist = dist ** 0.5
    return dist

def distance_cosine(a: str, b: str):
    # words_a = a.split(" ")
    # words_b = b.split(" ")
    # words_a = re.split(r",.\(\)!?'\"", a)
    # words_b = re.split(r",.\(\)!?'\"", b)
    # words_a = re.split(r",|.| |\(|\)|!|\?|'|\"", a)
    # words_b = re.split(r",|.| |\(|\)|!|\?|'|\"", b)
    words_a = re.split(r"[ ,.:;!?'\-()+\"\\<>]", a)
    words_b = re.split(r"[ ,.:;!?'\-()+\"\\<>]", b)
    # print(words_a)
    # print(words_b)

    for i in range(len(words_a)):
        words_a[i] = words_a[i].lower()
    for i in range(len(words_b)):
        words_b[i] = words_b[i].lower()

    all_words = set()
    for word in words_a:
        if word not in all_words:
            all_words.add(word)
    for word in words_b:
        if word not in all_words:
            all_words.add(word)
    dist = 0
    for word in all_words:
        temp = words_a.count(word) - words_b.count(word)
        dist += temp*temp

    max_dist = 0
    for word in all_words:
        temp = max(words_a.count(word), words_b.count(word))
        max_dist += temp*temp
    dist = dist / max_dist
    return dist


def judge_cluster_from_file(filename, separator, metric_func = distance_cosine):
    with open(filename, "r", encoding="UTF-8") as file:
        lines = file.readlines()
        # for XD in lines:
        #     print(XD[:-1])
    group_counter = 0
    summ = 0
    i = 0
    # counter = 0
    while i < len(lines):
        counter = 0
        start_range = i
        s = 0
        while lines[i][:-1] != separator:
            i += 1
        for k in range(start_range, i):
            if lines[k] == "\n":
                continue
            for l in range(k, i):
                if lines[l] == "\n":
                    continue
                # print(lines[k][:-1], "XDXDXD", lines[l][:-1])
                x = metric_func(lines[k][:-1], lines[l][:-1])
                # if not x:
                #     print(lines[k], lines[l])
                s += x
                counter += 1
        if counter != 1:
            s /= counter
            summ += s
            group_counter += 1
        # else:
        #     print("XD")
        i += 1
    return summ / group_counter

def most_common_words(filename):
    with open(filename, "r", encoding="UTF-8") as file:
        text = file.read()
        text = re.split(r",.!?'\"", text)
    word_counter = dict()
    for word in text:
        if not word_counter.get(word, 0):
            word_counter[word] = 0
        word_counter[word] += 1
    return word_counter

def clusterise(filename):
    with open(filename, "r", encoding="UTF-8") as file:
        lines = file.readlines()
    groups = []
    n = len(lines)
    counter = 0
    for line in lines:
        print(f"{counter} / {n}")
        counter += 1
        if line == "\n":
            continue
        for group in groups:
            if distance_cosine(group[0][:-1], line[:-1]) < 0.2:
                group.append(line)
                break
        else:
            groups.append([line])
    counter = 0
    with open("clustered.txt", "w", encoding="UTF-8") as file:
        for group in groups:
            for line in group:
                counter += 1
                file.write(line)
            file.write("\n")
            file.write("##########\n")
```
